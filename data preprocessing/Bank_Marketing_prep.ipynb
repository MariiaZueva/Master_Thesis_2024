```python
bm['default'].replace({'yes': True, 'no': False}, inplace=True)
bm['housing'].replace({'yes': True, 'no': False}, inplace=True)
bm['loan'].replace({'yes': True, 'no': False}, inplace=True)
bm['deposit'].replace({'yes': True, 'no': False}, inplace=True)
bm
```

```python
bm_encoded = pd.get_dummies(bm, columns=['job','marital','education', 'contact','month','poutcome'])
```

```python
bm_encoded[['job_admin.',
       'job_blue-collar', 'job_entrepreneur', 'job_housemaid',
       'job_management', 'job_retired', 'job_self-employed', 'job_services',
       'job_student', 'job_technician', 'job_unemployed', 'job_unknown',
       'marital_divorced', 'marital_married', 'marital_single',
       'education_primary', 'education_secondary', 'education_tertiary',
       'education_unknown', 'contact_cellular', 'contact_telephone',
       'contact_unknown', 'month_apr', 'month_aug', 'month_dec', 'month_feb',
       'month_jan', 'month_jul', 'month_jun', 'month_mar', 'month_may',
       'month_nov', 'month_oct', 'month_sep', 'poutcome_failure',
       'poutcome_other', 'poutcome_success', 'poutcome_unknown']].replace({1: True, 0: False}, inplace=True)

bm_encoded
```

```python
bm_binarized = pd.DataFrame()
bm_binarized['age_18-25'] = bm_encoded['age'] < 26
bm_binarized['age_26-40'] = (bm_encoded['age'] >=26) &  (bm_encoded['age'] <=40)
bm_binarized['age_41-65'] = (bm_encoded['age'] >40) &  (bm_encoded['age'] <=65)
bm_binarized['age_65+'] = bm_encoded['age'] >65
bm_binarized
```

```python
bm_binarized['pdays_0'] = bm_encoded['pdays'] == -1
bm_binarized['pdays_1-182'] = (bm_encoded['pdays'] >=0) &  (bm_encoded['pdays'] <183)
bm_binarized['pdays_183-365'] = (bm_encoded['pdays'] >=183) &  (bm_encoded['pdays'] <=365)
bm_binarized['pdays_366-730'] = (bm_encoded['pdays'] >365) & (bm_encoded['pdays'] <=730)
bm_binarized['pdays_731-854'] = (bm_encoded['pdays'] >730) & (bm_encoded['pdays'] <=854)
bm_binarized
```

```python
bm_binarized['balance<=0'] = bm_encoded['balance'] <=0
bm_binarized['0 < balance < 10000'] = (bm_encoded['balance'] > 0) &  (bm_encoded['balance'] <=10000)
bm_binarized['balance > 10000'] = bm_encoded['balance'] > 10000
```

```python
bm_binarized['day < 15'] = bm_encoded['day'] <15
bm_binarized['day >= 15'] = bm_encoded['day'] >= 15
```

```python
bm_binarized['campaign<=5'] = bm_encoded['campaign'] <=5
bm_binarized['campaign>5'] = bm_encoded['campaign'] > 5
```

```python
bm_binarized['previous<=5'] = bm_encoded['previous'] <=5
bm_binarized['previous>5'] = bm_encoded['previous'] > 5
```

```python
bm_encoded = bm_encoded.drop(['duration','pdays','age', 'balance', 'day', 'campaign', 'previous'], axis = 1)
```

```python
bm_encoded
```

```python
bm_encoded['default'].replace({'yes': True, 'no': False}, inplace=True)
bm_encoded['housing'].replace({'yes': True, 'no': False}, inplace=True)
bm_encoded['loan'].replace({'yes': True, 'no': False}, inplace=True)
bm_encoded['deposit'].replace({'yes': True, 'no': False}, inplace=True)
```

```python
bm_encoded =  pd.concat([bm_encoded, bm_binarized], axis=1)
```

```python
for col in bm_encoded.columns:
  print(col, ":", sorted(bm_encoded[col].unique()))
```

```python
bm_encoded['sample_id'] = np.arange(0, bm_encoded.shape[0], 1).astype(str)
bm_encoded['id'] = bm_encoded['sample_id'].apply(lambda x:  'h'+x)
bm_encoded.drop(columns=['sample_id'], inplace=True)
bm_encoded=bm_encoded.set_index('id')
bm_encoded
```
